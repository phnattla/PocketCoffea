```python
import os
from coffea.util import load
import hist
```

```python
folder = "/work/mmarcheg/PocketCoffea/output/datamc/ttHbb_ttbar_ST_datamc_btagcalibrated_v02"
filename = list(filter(lambda x : 'dataset' in x, os.listdir(folder)))[0]
filename
```

```python
o = load(os.path.join(folder, filename))
o.keys()
```

```python
h = o['variables']["JetGood_eta_1"]
variations = list(h['ttHTobb'].axes[1])
for v in variations:
    samples_mc = [s for s in h.keys() if not 'DATA' in s]
    slicing_mc = {'cat' : 'SingleMuon_1b', 'variation' : v, 'year' : '2018'}
    h_mc_dict = {s : h[s][slicing_mc] for s in samples_mc}
    #h_mc_dict.keys()
    stack_mc = hist.Stack.from_dict(h_mc_dict)
    print("****************")
    print(v)
    stack_mc.show()
```
